# Student Blog




## Date:
9/9/25




### Summary:
We discussed our ideas, divided tasks, and set clear goals for the day.
- After finalizing our plan, we focused on getting the spaceship game up and running, collaborating to troubleshoot issues and implement new features.
- Once we made significant progress on the game, we documented our experiences and reflections by creating this blog to track our journey and share what we learned.








### What I Learned Today
- The importance of team collaboration and clear communication when working on group projects.
- How to effectively strategize and outline an approach before starting a task.
- The value of dividing tasks and setting clear goals to ensure productivity.
- Practical experience in troubleshooting and implementing new features in a project.
- The benefits of documenting progress and reflections to track learning and share insights.




### Challenges I Faced
- Integrating new features into the existing codebase without introducing bugs.
- Resolving merge conflicts when collaborating with teammates on the same files.
- Understanding unfamiliar code sections written by other team members.
- Managing time effectively to balance debugging and feature development.
- Ensuring that all team members were on the same page regarding project goals and progress.


### Me vs The Begining of the year
At the start of this program, I was painfully shy and felt completely lost, having come in three days late with zero coding knowledge. This immediately put me at a disadvantage, and the stress of not being able to contribute or understand the material was overwhelming. Now, the overwhelming fear of coding is gone; I approach problems with confidence and the knowledge that I can build things. This transformation was a direct result of forcing myself to step out of my shell and learning to openly communicate with my team. My ability to collaborate and ask targeted questions is the single biggest change I've experienced.


### Key Takeaways
The three sprints provided a roller coaster of challenges and victories. In Sprint 1 (Tools), I focused heavily on soft skills, learning that asking questions and being open about what I didn't know was the only way to catch up. Sprint 2 (Fundamentals of JavaScript) was my lowest point; significant issues at home and a lack of internet access severely hampered my ability to complete assignments, resulting in a very low score. Sprint 3, however, was my best: I truly "locked in," working independently to design all the game sprites, develop the initial Quest of Spook ideation, and successfully integrate complex mechanics like shooting and enemy death.


### Night at the Museum
The Night at the Museum event was a fantastic experience that validated all our hard work. Seeing people genuinely interact with our game was incredibly rewarding and a proud moment for the entire team. I specifically loved the comments we received about the game's overall concept and smooth mechanics. The best piece of personal feedback was a compliment on the cohesion and professional look of the sprites I had created, which made the late nights of pixel work totally worth it. It felt great to have my creative contributions, alongside the code, recognized by an audience.

### What to do next time
If our team were given another week to work on the Quest of Spook project, my immediate focus would be on adding a robust layer of polish and essential player tracking. The top priority would be implementing a fully functional scoring system with a persistent high-score leaderboard to incentivize replayability. We also need to add subtle, but crucial, elements of "game feel," such as a distinct visual flash or sound effect when the player takes damage. These small additions would significantly improve the user experience and make the game feel more complete and professional.


### What next?
Given how much I enjoyed the creative and functional challenges of building a game, I want to formally pursue the visual side of computer science next. My plan is to start learning basic 3D modeling and animation to understand how complex game assets are truly created. This involves getting comfortable with software like Blender and learning how to export and integrate assets into a game engine like Unity. This step feels like the natural evolution of the sprite work I enjoyed in Sprint 3, moving from 2D pixel art to 3D environment and character design.
